```python
tensor = torch.arange(1,100)
```

```python
%%time
def multi(mat1,mat2):
    sum = 0 
    for i in range(len(mat1)):
        sum += mat1[i] * mat2[i]
    
    return sum

multi(tensor,tensor)
```

```python
%%time
torch.matmul(tensor,tensor)
```
